```python
import pandas as pd
df=pd.read_csv('fake_or_real_news.csv')
```

```python

df.head()
```

```python
df.shape
```

```python
df['text'][4]
```

```python
# prompt: drop id

df.drop('id', axis=1, inplace=True)
df.head()
```

```python
x = df[['text', 'title']]  # Pass column names as a list
y = df['label']
```

```python
# convert y to binary
y = y.map({'FAKE': 0, 'REAL': 1})
```

```python
# splitting the dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
x_train_vec=vectorizer.fit_transform(x_train['title']+x_train['text'])
x_test_vec=vectorizer.transform(x_test['title']+x_test['text'])
```

```python
# prompt: use 'word2vec-google-news-300' for thid dataset

from sklearn.linear_model import LogisticRegression

# Initialize and train the logistic regression model
model = LogisticRegression(max_iter=1000)  # Increased max_iter
model.fit(x_train_vec, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test_vec)

# Evaluate the model (example: accuracy)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# prompt: reort and evluate

from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# ... (Your existing code) ...

# Make predictions on the test set
y_pred = model.predict(x_test_vec)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Generate and print the classification report
print(classification_report(y_test, y_pred))

# Generate and display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=['FAKE', 'REAL'], yticklabels=['FAKE', 'REAL'])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```
